```python
import pandas as pd
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import os
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import cross_val_score, train_test_split
```

```python
df = pd.read_csv('kc_house_data_NaN.csv')
df.head(10)
```

```python
df.drop(['Unnamed: 0', 'id'], axis = 1,  inplace = True)
```

```python
df.describe()
```

```python
print("number of NaN values for the column bedrooms :", df['bedrooms'].isnull().sum())
print("number of NaN values for the column bathrooms :", df['bathrooms'].isnull().sum())
```

```python
mean_bedrooms = df['bedrooms'].mean()
mean_bathrooms = df['bathrooms'].mean()

df['bedrooms'] = df['bedrooms'].fillna(mean_bedrooms)
df['bathrooms'] = df['bathrooms'].fillna(mean_bathrooms)
```

```python
print("number of NaN values for the column bedrooms :", df['bedrooms'].isnull().sum())
print("number of NaN values for the column bathrooms :", df['bathrooms'].isnull().sum())
```

```python
df_numeric = df.select_dtypes(include=[np.number])
df_numeric.corr()['price'].sort_values()
```

```python
X = df[['long']]
Y = df['price']
lm = LinearRegression()
lm.fit(X,Y)
lm.score(X, Y)
```

```python
X = df[['sqft_living']]
lm.fit(X,Y)
lm.score(X,Y)
```

```python
Z = df.drop(columns=['date', 'price'], errors='ignore')
```

```python
lm.fit(Z,Y)
lm.score(Z,Y)
```

```python
Input=[('scale',StandardScaler()), ('polynomial', PolynomialFeatures(include_bias=False)), ('model', LinearRegression())]
pipe=Pipeline(Input)
Z = Z.astype(float)
pipe.fit(Z,Y)
ypipe=pipe.predict(Z)
print(r2_score(Y,ypipe))
```

```python
x_train, x_test, y_train, y_test = train_test_split(Z, Y, test_size=0.15, random_state=1)
print("number of test samples:", x_test.shape[0])
print("number of training samples:",x_train.shape[0])
```

```python
ridge=Ridge(alpha=0.1)
ridge.fit(x_train,y_train)
ridge.score(x_test,y_test)
```

```python
pr = PolynomialFeatures(degree=2)
x_train_pr = pr.fit_transform(x_train)
x_test_pr = pr.transform(x_test)
scaler = StandardScaler()
x_train_pr_scaled = scaler.fit_transform(x_train_pr)
x_test_pr_scaled = scaler.transform(x_test_pr)
ridge = Ridge(alpha=0.1)
ridge.fit(x_train_pr_scaled, y_train)
y_hat = ridge.predict(x_test_pr_scaled)
print(r2_score(y_test, y_hat))
```
